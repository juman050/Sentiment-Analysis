```python
import pandas as pd
from matplotlib import pyplot as plt

# Data for the comparison
comparison_data = {
    "Model": ["Logistic Regression (Chosen)", "Logistic Regression (Model Project)", "SVM (Chosen)"],
    "Accuracy": ["78.0%", "75.0%", "77.0%"],
    "Precision (Negative)": ["79%", "76%", "78%"],
    "Recall (Negative)": ["76%", "72%", "75%"],
    "F1 Score (Negative)": ["77%", "73%", "76%"],
    "Precision (Positive)": ["77%", "73%", "76%"],
    "Recall (Positive)": ["80%", "77%", "79%"],
    "F1 Score (Positive)": ["78%", "75%", "78%"]
}

# Creating the dataframe
comparison_df = pd.DataFrame(comparison_data)

# Displaying the dataframe
comparison_df
```

```python
# Updated data for the chart
model_project_updated = {
    "Accuracy": 75.0,
    "Precision (Negative)": 76.0,
    "Recall (Negative)": 72.0,
    "F1 Score (Negative)": 73.0,
    "Precision (Positive)": 73.0,
    "Recall (Positive)": 77.0,
    "F1 Score (Positive)": 75.0
}

proposed_svm_updated = {
    "Accuracy": 77.0,
    "Precision (Negative)": 78.0,
    "Recall (Negative)": 75.0,
    "F1 Score (Negative)": 76.0,
    "Precision (Positive)": 76.0,
    "Recall (Positive)": 79.0,
    "F1 Score (Positive)": 78.0
}

proposed_lr_updated = {
    "Accuracy": 78.0,
    "Precision (Negative)": 79.0,
    "Recall (Negative)": 76.0,
    "F1 Score (Negative)": 77.0,
    "Precision (Positive)": 77.0,
    "Recall (Positive)": 80.0,
    "F1 Score (Positive)": 78.0
}

labels_updated = list(model_project_updated.keys())
model_project_values_updated = list(model_project_updated.values())
proposed_svm_values_updated = list(proposed_svm_updated.values())
proposed_lr_values_updated = list(proposed_lr_updated.values())

x_updated = range(len(labels_updated))

plt.figure(figsize=(14, 8))
plt.plot(x_updated, model_project_values_updated, label="Model Project (LR)", marker='o')
plt.plot(x_updated, proposed_svm_values_updated, label="Chosen Approach (SVM)", marker='o')
plt.plot(x_updated, proposed_lr_values_updated, label="Chosen Approach (LR)", marker='o')

plt.xticks(x_updated, labels_updated, rotation=45)
plt.xlabel('Metrics')
plt.ylabel('Percentage')
plt.title('Comparison of Model Project and Chosen Approach')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import numpy as np

# Updated data
labels_updated = list(model_project_updated.keys())
model_project_values_updated = list(model_project_updated.values())
proposed_svm_values_updated = list(proposed_svm_updated.values())
proposed_lr_values_updated = list(proposed_lr_updated.values())

x_updated = np.arange(len(labels_updated))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(x_updated - width, model_project_values_updated, width, label='Model Project (LR)', color='blue')
rects2 = ax.bar(x_updated, proposed_svm_values_updated, width, label='Chosen Approach (SVM)', color='green')
rects3 = ax.bar(x_updated + width, proposed_lr_values_updated, width, label='Chosen Approach (LR)', color='yellow')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Metrics')
ax.set_ylabel('Percentage')
ax.set_title('Comparison of Model Project and Chosen Approach')
ax.set_xticks(x_updated)
ax.set_xticklabels(labels_updated, rotation=45)
ax.legend()

# Adding grid
ax.grid(True)

# Adding the value labels on top of the bars
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}%',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(rects1)
add_labels(rects2)
add_labels(rects3)

fig.tight_layout()

plt.show()
```

#initial confusion matrix of LR and SVM ✅

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Data for confusion matrices
confusion_matrix_lr = np.array([[179725, 59636], [49920, 190719]])
confusion_matrix_svm = np.array([[178513, 60848], [49549, 191090]])

# Plotting the confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(confusion_matrix_lr, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('Logistic Regression Confusion Matrix')
axes[0].set_xlabel('Predicted Labels')
axes[0].set_ylabel('True Labels')
axes[0].xaxis.set_ticklabels(['Negative', 'Positive'])
axes[0].yaxis.set_ticklabels(['Negative', 'Positive'])

sns.heatmap(confusion_matrix_svm, annot=True, fmt='d', cmap='Blues', ax=axes[1])
axes[1].set_title('SVM Confusion Matrix')
axes[1].set_xlabel('Predicted Labels')
axes[1].set_ylabel('True Labels')
axes[1].xaxis.set_ticklabels(['Negative', 'Positive'])
axes[1].yaxis.set_ticklabels(['Negative', 'Positive'])

plt.tight_layout()
plt.show()
```
